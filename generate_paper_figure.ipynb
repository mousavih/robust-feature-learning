# Run Experiments

In this section, we run 6 different experiment settings as described in the paper. For reproducibility, at the beginning of each experinemt, we set the random number generator seed to 42.

```python
# Set hyperparameters

N_experiments = 3
N = 100
N_true = 1
d = 100
batch_size = 300
num_test_samples = 10000
epsilon = 1.0
```

```python
all_costs_adv, all_costs_init_fixed, all_costs_init_trained = {}, {}, {}
```

```python
def initialize_experiment(seed):
    torch.manual_seed(seed)
    
    W_true = torch.randn(N_true, d, device=device)
    W_true /= torch.linalg.norm(W_true, dim=1).reshape(-1, 1)
    a_true = 1 / np.sqrt(N_true) * torch.ones(N_true, device=device)
    b_true = torch.zeros(N_true, device=device)
    
    return W_true, a_true, b_true
```

## ReLU (Known Direction)

```python
W_true, a_true, b_true = initialize_experiment(seed=42)

costs_adv, costs_init_fixed, costs_init_trained = run_experiment_known_dir(N_experiments=N_experiments,
                                                                                    N=N, N_true=N_true, 
                                                                                    d=d, batch_size=batch_size,
                                                                                    epsilon=epsilon,
                                                                                    num_test_samples=num_test_samples,
                                                                                    W_true=W_true, a_true=a_true, b_true=b_true,
                                                                                    target_func=F.relu)

all_costs_adv['relu_known'] = costs_adv
all_costs_init_fixed['relu_known'] = costs_init_fixed
all_costs_init_trained['relu_known'] = costs_init_trained
```

# Plots

Here, we generate the figure seen in the paper.

```python
# If the results are saved, they can be loaded from the given address.
def get_vals(target, known_dir, load_from_file=False):
    if load_from_file:
        if known_dir:
            prefix = f'results/{target}_known_costs_'
        else:
            prefix = f'results/{target}_unknown_costs_'
            
        
        costs_adv = torch.load(prefix + 'adv.pt')
        costs_init_fixed = torch.load(prefix + 'init_fixed.pt')
        costs_init_trained = torch.load(prefix + 'init_trained.pt')
    else:        
        costs_adv = all_costs_adv[target + '_known'] if known_dir else all_costs_adv[target + '_unknown']
        costs_init_fixed = all_costs_init_fixed[target + '_known'] if known_dir else all_costs_init_fixed[target + '_unknown']
        costs_init_trained = all_costs_init_trained[target + '_known'] if known_dir else all_costs_init_trained[target + '_unknown']
    
    adv_mean = costs_adv.mean(dim=0)
    adv_std = costs_adv.std(dim=0)

    init_fixed_mean = costs_init_fixed.mean(dim=0)
    init_fixed_std = costs_init_fixed.std(dim=0)

    init_trained_mean = costs_init_trained.mean(dim=0)
    init_trained_std = costs_init_trained.std(dim=0)
    
    return adv_mean, adv_std, init_fixed_mean, init_fixed_std, init_trained_mean, init_trained_std
```

```python
from matplotlib.ticker import MaxNLocator


targets = ['relu', 'tanh', 'he2']
target_names = ['ReLU', 'Tanh', 'He2']

color1 = '#e41a1c'
color2 = '#377eb8'
color3 = '#4daf4a'
color4 = '#984ea3'
color5 = '#ff7f00'

def generate_plot(load_from_file, savefig):
    tick_pos = [i * 50 for i in range(100)]
    fig, axes = plt.subplots(2, 3, figsize=(12 / 1.2, 8 / 1.2))

    for i in range(3):
        axes[0][i].yaxis.set_major_locator(MaxNLocator(nbins=5))
        axes[0][i].xaxis.set_major_locator(MaxNLocator(nbins=4))
        axes[0][i].set_title(f'Teacher: {target_names[i]} (Known Direction)')
        axes[0][i].set_xlabel('Iterations')

        adv_mean, adv_std, init_fixed_mean, init_fixed_std, init_trained_mean, init_trained_std = get_vals(targets[i], True, load_from_file)

        axes[0][i].plot(tick_pos, adv_mean, label='Full AD training', linewidth=2, color=color1)
        axes[0][i].fill_between(tick_pos, adv_mean - adv_std, adv_mean + adv_std, color=color1, alpha=0.3)

        axes[0][i].plot(tick_pos, init_fixed_mean, label='$W$ fixed at target $u$', linewidth=2, color=color2)
        axes[0][i].fill_between(tick_pos, init_fixed_mean - init_fixed_std, init_fixed_mean + init_fixed_std, 
                             color=color2, alpha=0.3)

        axes[0][i].plot(tick_pos, init_trained_mean, label='$W$ init from $u$', linewidth=2, color=color3)
        axes[0][i].fill_between(tick_pos, init_trained_mean - init_trained_std, init_trained_mean + init_trained_std, 
                             color=color3, alpha=0.3)


    for i in range(3):
        if i == 2:
            tick_pos = [i * 70 for i in range(100)]

        axes[1][i].yaxis.set_major_locator(MaxNLocator(nbins=4))
        axes[1][i].xaxis.set_major_locator(MaxNLocator(nbins=4))
        axes[1][i].set_title(f'Teacher: {target_names[i]} (Unknown Direction)')
        axes[1][i].set_xlabel('Iterations')

        adv_mean, adv_std, init_fixed_mean, init_fixed_std, init_trained_mean, init_trained_std = get_vals(targets[i], False, load_from_file)

        axes[1][i].plot(tick_pos, adv_mean, label='Full AD training', linewidth=2, color=color1)
        axes[1][i].fill_between(tick_pos, adv_mean - adv_std, adv_mean + adv_std, color=color1, alpha=0.3)

        axes[1][i].plot(tick_pos, init_fixed_mean, label='$W$ fixed at SD training', linewidth=2, color=color4)
        axes[1][i].fill_between(tick_pos, init_fixed_mean - init_fixed_std, init_fixed_mean + init_fixed_std, 
                             color=color4, alpha=0.3)

        axes[1][i].plot(tick_pos, init_trained_mean, label='$W$ init from SD training', linewidth=2, color=color5)
        axes[1][i].fill_between(tick_pos, init_trained_mean - init_trained_std, init_trained_mean + init_trained_std, 
                             color=color5, alpha=0.3)


    axes[0][0].set_ylabel('Robust Test Risk')
    axes[1][0].set_ylabel('Robust Test Risk')

    handles1, labels1 = axes[0][0].get_legend_handles_labels()
    handles2, labels2 = axes[1][0].get_legend_handles_labels()

    axes[0, 0].legend()

    axes[1, 0].legend()


    plt.tight_layout(pad=2)

    if savefig:
        plt.savefig('experiments.pdf')
    plt.show()
```
